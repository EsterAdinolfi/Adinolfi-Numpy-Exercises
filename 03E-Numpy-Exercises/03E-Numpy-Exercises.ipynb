## Base Practice

#### 1. Create a null vector of size 10 but the fifth value which is 1 (base)
`hint: array[4]`

```python
vector = np.zeros(10, dtype=int)
vector[4]=1

print(vector)
```

#### 2. Create a 3x3 matrix with values ranging from 0 to 8 with zeros on the diagonal (base)
`hint: np.arange, reshape, np.eye`

```python
vector = np.arange(9, dtype=int)
matrix = vector.reshape(3,3)
matrix = matrix - np.eye(3)*matrix
matrix = matrix.astype(int)

print(matrix)
```

#### 3. How to add a border (filled with 0's) around an 5x5 matrix of ones? (base)
`hint: np.ones, np.pad`

```python
matrix = np.ones((5,5), dtype=int)
matrix = np.pad(matrix, 1, 'constant')

print(matrix)
```

#### 4. Normalize a 5x5 random matrix (base)
`hint: (x - mean) / std, np.mean, np.std`

```python
matrix = np.random.rand(5, 5)
mean = np.mean(matrix)
std = np.std(matrix)

normalized_matrix = (matrix - mean) / std

print("Original matrix is:")
print(matrix)
print("Normalized matrix is:")
print(normalized_matrix)
```

#### 5. Multiply a 5x3 matrix of ones by a 3x2 matrix of ones (real matrix product) (base)
`hint: np.dot`

```python
matrix1 = np.ones((5,3), dtype=int)
matrix2 = np.ones((3,2), dtype=int)

result = np.dot(matrix1,matrix2)

print (result)
```

#### 7. Create a random vector of size 10 with values in range (-3, 12) and sort it (intermediate)
`hint: sort`

```python
vector = np.random.randint(-3, 12, 10)
print('Random vector:')
print(vector)

vector.sort()
print('Vector after sorting:')
print(vector)
```

#### 8. Create random vector of size 10 and replace the maximum value by its additive inverse and the minimum value (of the original array) with the median value (intermediate)
`hint: argmax, argmin`

```python
vector = np.random.randint(-3, 12, 10)
print('Random vector:')
print(vector)

new_vector = vector.copy()

max_index = np.argmax(vector)
min_index = np.argmin(vector)

new_vector[max_index] = -vector[max_index]
new_vector[min_index] = np.median(vector)

print('New vector:')
print(new_vector)
```

#### 9. Randomly replace p elements in a 2D nxn zero matrix to 1 (intermediate)
`hint: np.put, np.random.choice`

```python
n = 5  # Size of the matrix (nxn)
p = 8  # Number of elements to replace

zero_matrix = np.zeros((n, n), dtype=int)

total_elements = n * n
p = min(p, total_elements)
indices_to_replace = np.random.choice(total_elements, p, replace=False)
np.put(zero_matrix, indices_to_replace, 1)

print("Modified Matrix:")
print(zero_matrix)
```

#### 10. Subtract the mean of each row of a random 5x10 matrix (intermediate)
`hint: mean(axis=,keepdims=)`

```python
random_matrix = np.random.rand(5, 10)
print('Initial random matrix:')
print(random_matrix)

row_means = np.mean(random_matrix, axis=1, keepdims=True)
new_matrix = random_matrix - row_means

print("New matrix:")
print(new_matrix)
```
